```python
class Person:
    def __init__(self,Name,Branch,Theme,NameOfPaper):
        self.Name=Name
        self.Branch=Branch
        self.Theme=Theme
        self.NameOfPaper=NameOfPaper
    def display(self):
        print("Name of Researcher : ",self.Name)
        print("Name of Branch : ",self.Branch)
        print("Name of Theme : ",self.Theme)
        print("Name of Research Topic : ",self.NameOfPaper)
class Institute(Person):
    def __init__(self,Name,Branch,Theme,NameOfPaper,instName,totalCourses):
        super().__init__(Name,Branch,Theme,NameOfPaper)
        self.instName=instName
        self.totalCourses=totalCourses
    def display(self):
        Person.display(self)
        print("Name of institute ",self.instName)
        print("Total courses : \n",[i for i in self.totalCourses])
class UniversityData(Institute):
    def __init__(self,Name,Branch,Theme,NameOfPaper,instName,totalCourses,uniName,specs):
        super().__init__(Name,Branch,Theme,NameOfPaper,instName,totalCourses)
        self.uniName=uniName
        self.specs=specs
    def display(self):
        Institute.display(self)
        print("List of INSTITUTes : \n",[i for i in self.uniName],"\nTheir Specialization Respectively : \n",[i for i in self.specs])
```

# Operator OverLoading

```python
class OP:
    def __init__(self,x):
        self.x=x
    def __add__(self,A):
        # A Creates a object and Automatically Stores value of next or previous Operation
        return self.x-A.x
```

```python
class Complex:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __add__(self,other):
        Temp=Complex(self.x,self.y)
        Temp.x=self.x+other.x
        Temp.y=self.y+other.y
        return Temp
    def display(self):
        print(f"{self.x}+{self.y}j")
```

```python
# write a pogram to fetch integer value as an object comapre two objects and show whether it is equal to , greater, or less than other object
class Comp:
    def __init__(self,x):
        self.x=x
    def __eq__(self,A):
        flag=False
        if self.x==A.x:
            flag=True
        return flag
    def __gt__(self,A):
        flag=False
        if self.x>A.x:
            flag=True
        return flag
    def __lt__(self,A):
        flag=False
        if self.x<A.x:
            flag=True
        return flag
```

```python
# WAP CLASS TO COMAPRE DATES
class Compare:
    def __init__(self,day,month,year):
        self.day=day
        self.month=month
        self.year=year
    def __eq__(self,A):
        if self.day==A.day and self.month==A.month and self.year==A.year:
            return True
        else:
            return False
    def __gt__(self,A):
        if self.year>A.year:
            return True
        else:
            if self.month>A.month:
                return True
            else:
                if self.day>A.day:
                    return True
                else:
                    return False
    def __lt__(self,A):
        if self.year<A.year:
            return True
        else:
            if self.month<A.month:
                return True
            else:
                if self.day<A.day:
                    return True
                else:
                    return False
```

```python
d1=Compare(1,2,2026)
d2=Compare(5,2,2026)
```

```python
d1<d2
```
