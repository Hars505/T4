## WAP TO create research database Which has 3 classes named 'person','institude' which inherits the person and University data which inherits the institude all have a common method named display to print statistics the class person collects data like name,branch,theme,name of paper class institute collects data like name of institude, and total courses avaliable and class University data collects data like list of institutes and their specialization 

```python
class Person:
    def __init__(self,Name,Branch,Theme,NameOfPaper):
        self.Name=Name
        self.Branch=Branch
        self.Theme=Theme
        self.NameOfPaper=NameOfPaper
    def display(self):
        print("Name of Researcher : ",self.Name)
        print("Name of Branch : ",self.Branch)
        print("Name of Theme : ",self.Theme)
        print("Name of Research Topic : ",self.NameOfPaper)
class Institute(Person):
    def __init__(self,Name,Branch,Theme,NameOfPaper,instName,totalCourses):
        super().__init__(Name,Branch,Theme,NameOfPaper)
        self.instName=instName
        self.totalCourses=totalCourses
    def display(self):
        Person.display(self)
        print("Name of institute ",self.instName)
        print("Total courses : \n",[i for i in self.totalCourses])
class UniversityData(Institute):
    def __init__(self,Name,Branch,Theme,NameOfPaper,instName,totalCourses,uniName,specs):
        super().__init__(Name,Branch,Theme,NameOfPaper,instName,totalCourses)
        self.uniName=uniName
        self.specs=specs
    def display(self):
        Institute.display(self)
        print("List of INSTITUTes : \n",[i for i in self.uniName],"\nTheir Specialization Respectively : \n",[i for i in self.specs])
```

```python
class Person:
    def __init__(self,Name,Branch,Theme,NameOfPaper):
        self.Name=Name
        self.Branch=Branch
        self.Theme=Theme
        self.NameOfPaper=NameOfPaper
    def display(self):
        print("Name of Researcher : ",self.Name)
        print("Name of Branch : ",self.Branch)
        print("Name of Theme : ",self.Theme)
        print("Name of Research Topic : ",self.NameOfPaper)
class Institute:
    def __init__(self,instName,totalCourses):
        self.instName=instName
        self.totalCourses=totalCourses
    def display(self):
        print("Name of institute ",self.instName)
        print("Total courses : \n"+"\n".join(" -- "+i for i in self.totalCourses))
class UniversityData(Person,Institute):
    def __init__(self,Name,Branch,Theme,NameOfPaper,instName,totalCourses,uniName,specs):
        Person.__init__(self,Name,Branch,Theme,NameOfPaper)
        Institute.__init__(self,instName,totalCourses)
        self.uniName=uniName
        self.specs=specs
    def display(self):
        Person.display(self)
        Institute.display(self)
        print("List of INSTITUTes : \n"+"\n".join(" -- "+i for i in self.uniName),"\nTheir Specialization Respectively : \n"+"\n".join(" -- "+i for i in self.specs))
```

# Operator OverLoading

```python
class OP:
    def __init__(self,x):
        self.x=x
    def __add__(self,A):
        # A Creates a object and Automatically Stores value of next or previous Operation
        return self.x-A.x
```
