#### WAP to create a line chart  showing marks of 4 different subjects for four different exams(t1 to t4) the subjects are Python - orange(circle),FSD GREEN(star), DE - MAGENTA(DIAMOND),PS BLACK(point) give proper labelling and titles

```python
Python = np.array([12,18,17,24])
FSD = np.array([14,18,15,20])
DE = np.array([24,20,22,23])
PS = np.array([16,17,19,21])
exam=np.array(["T1","T2","T3","T4"])
subject=np.array(["python","FSD","DE","PS"])
font1={"family":"serif","color":"Blue","size":20}
font2={"family":"sans-serif","color":"RED","size":20}
font3={"family":"Monospace","color":"Blue","size":20}
font4={"family":"cursive","color":"RED","size":20}
font5={"family":"fantasy","color":"RED","size":20}
plt.plot(exam,Python,mfc="orange",marker=".",ms=10,linewidth=2)
plt.plot(exam,FSD,mfc="green",marker="*",ms=10,linewidth=2)
plt.plot(exam,DE,mfc="m",marker="D",ms=10,linewidth=2)
plt.plot(exam,PS,"o--k",ms=10,linewidth=2)
plt.xlabel("Exams",fontdict=font1)
plt.ylabel("Marks",fontdict=font4)
plt.title("Marks of student",fontdict=font5)
plt.legend(subject,loc=0)
plt.grid()
plt.savefig("Marks.jpg")
plt.show()
```

### WAP TO CREATE A SCATTER PLOT WHICH HAS 3 DIFFERENT GROUPING 1ST GROUP HAS 50 VALUES ALL BETWEEN 0 AND 1 , 2nd GROUP HAS 50 VALUES BETWEEN 1 to 50, 3rd GROUPING HAS VALUES ALL ABOVE 50 

```python
x=np.random.rand(50)
y=np.random.randint(50,size=50)
z=np.random.randint(50,100,size=50)
color=np.array([i*50 for i in range(50)])
plt.scatter(x,y,z,c=color,cmap="magma",alpha=0.5)
plt.show()
```

## AREA PLOTS 

```python
x=np.arange(4)
y=np.array([20,30,25,25])
plt.plot(x,y,color="black")
plt.fill_between(x,y,color="SKYBLUE",alpha=0.5)
plt.show()
```

```python
income=np.array([2,3,8,7,6,9,4,3,5,4,6,5])
expense=np.array([1,5,7,3,8,8,2,1,4,9,3,2])
month=np.arange(12)
plt.figure(figsize=(8,8))
plt.plot(month,income,color="Green")
plt.plot(month,expense,color="red")
plt.fill_between(month,income,expense,color="green",where=income>expense,interpolate=True,alpha=0.5,label="Profit")
plt.fill_between(month,income,expense,color="red",where=income<=expense,interpolate=True,alpha=0.5,label="Loss")
plt.legend()
plt.show()
```
